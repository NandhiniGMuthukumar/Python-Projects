# Wikipedia: List of largest Australian companies


```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_Australian_companies'
```

```python
response = requests.get(url)
```

```python
result = BeautifulSoup(response.text, 'html.parser')
```

```python
table_1 = result.find('table')
table_2 = result.find_all('table')[1]
```

```python
table_1
```

```python
table_data_1 = []
for row in table_1.find_all('tr'):
    row_data = [cell.text.strip() for cell in row.find_all(['th', 'td'])]
    table_data_1.append(row_data)
table_data_2 = []
for row in table_2.find_all('tr'):
    row_data = [cell.text.strip() for cell in row.find_all(['th', 'td'])]
    table_data_2.append(row_data)
```

```python
for row in table_data_1:
    print(row)
```

```python
for row in table_data_2:
    print(row)
```

```python
import pandas as pd
```

```python
column_name = table_data_1[0]
df_1 = pd.DataFrame(table_data_1[1:], columns=column_name)
column_name = table_data_2[0]
df_2 = pd.DataFrame(table_data_2[1:], columns=column_name)
```
